# Methylation Weight
| 优先级 | 规则代码 | 判定条件 (Condition) | 逻辑描述 | 建议权重 ($w$) |
| :--- | :--- | :--- | :--- | :--- |
| **L1** | **Annotation Hit** | `Gene_Annotation` 非空 | 注释文件（如图片中的 Gene_Annotation 列）有明确基因名。 | **1.0** |
| **L2** | **RAG Validated** | 注释为空 + RAG验证为真 | 注释为空，但通过 RAG 检索发现 TSS 临近基因有文献支持调控关系。 | **1.0** |
| **L3** | **Enhancer** | `Status` = "Enhancer" | 注释为空，RAG 未验证，但 Regulatory_Status 包含 "Enhancer"。 | **0.8** |
| **L4** | **Open Chromatin** | `Status` = "DHS"/"OpenChromatin" | 注释为空，RAG 未验证，但 Regulatory_Status 为 "OpenChromatin" 或 "DHS"。 | **0.7** |
| **L5** | **Proximal** | Dist < 2kb (或其他阈值) | 注释为空，无功能注释，但距离 TSS 较近（即近端调控）。 | **0.5** |
| **L6** | **Distal / Other** | `Status` = "Distal" / "Potential_Inactive" | 注释为空，且标注为 "Distal" 或 "Potential_Inactive"。 | **0.3** |
